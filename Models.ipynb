```python
import pandas as pd
data = pd.read_csv('ts_data.csv')
data.head()
```

```python
data.info()
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)
```

```python
data.to_csv('ts_data.csv')
```

```python
data.Failure = data.Failure.replace(['No', 'Yes'], [0,1])
```

```python
data.to_csv('ts_data_target_numerical.csv')
```

```python
data = pd.get_dummies(data, columns=['Operator', 'Measure2', 'Measure3'])
```

```python
data.head()
```

```python
data.to_csv('ts_data_target_numerical_onehotencoded.csv')
```

```python
features = data.drop('Failure', axis=1)
target = data.Failure
```

```python
f_data = pd.read_csv('filtered_measurements.csv')
```

```python
f_data.head()
```

```python
f_data.reset_index(inplace=True)
```

```python
f_data.info()
```

```python
f_data.drop(['Unnamed: 0', 'index'], axis=1, inplace=True)
```

```python
f_data.info()
```

```python
data2 = data.drop(['Measure1','Measure4','Measure5','Measure6','Measure7',\
                       'Measure8','Measure9','Measure10','Measure11','Measure12',\
                       'Measure13','Measure14','Measure15'], axis=1)
```

```python
data2.info()
```

```python
data2.reset_index(inplace=True)
```

```python
merge = pd.concat([data2,f_data], axis=1)
```

```python
merge.head()
```

```python
merge.to_csv('merged_filtered_data.csv')
```

```python
features1 = merge.drop(['Failure','Date'], axis=1)
target1 = merge.Failure
```

```python
features1.info()
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
import xgboost as xgb
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report
scaler = StandardScaler()
scaled_df = pd.DataFrame(scaler.fit_transform(features), columns=features.columns)
scaled_df.head()
##### set train, test
X_train, X_test, y_train, y_test = train_test_split(scaled_df, target, test_size=0.25)
###### create xgb instance with parameters
clf = xgb.XGBClassifier()
###### fit data
clf.fit(X_train, y_train)
###### make predictions
training_preds = clf.predict(X_train)
val_preds = clf.predict(X_test)
###### run classification report and confusion matrix
gbt_confusion_matrix = confusion_matrix(y_test, val_preds)
gbt_confusion_matrix
gbt_classification_report = classification_report(y_test, val_preds)
print(gbt_classification_report)
###### cross validation score
print('Mean XGB Cross-Val Score (k=5):')
print(cross_val_score(clf, scaled_df, target, cv=5).mean())
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
import xgboost as xgb
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report
scaler = StandardScaler()
scaled_df = pd.DataFrame(scaler.fit_transform(features1), columns=features1.columns)
scaled_df.head()
##### set train, test
X_train, X_test, y_train, y_test = train_test_split(scaled_df, target1, test_size=0.25)
###### create xgb instance with parameters
clf = xgb.XGBClassifier()
###### fit data
clf.fit(X_train, y_train)
###### make predictions
training_preds = clf.predict(X_train)
val_preds = clf.predict(X_test)
###### run classification report and confusion matrix
gbt_confusion_matrix = confusion_matrix(y_test, val_preds)
gbt_confusion_matrix
gbt_classification_report = classification_report(y_test, val_preds)
print(gbt_classification_report)
###### cross validation score
print('Mean XGB Cross-Val Score (k=5):')
print(cross_val_score(clf, scaled_df, target, cv=5).mean())
```
