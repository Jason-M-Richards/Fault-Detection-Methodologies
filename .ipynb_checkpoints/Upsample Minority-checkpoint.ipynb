```python
import pandas as pd
```

```python
data = pd.read_csv('ts_data.csv')
data.head()
```

```python
features = data.drop('Failure', axis=1)
target = data.Failure
```

```python
#upsample minority
from imblearn.over_sampling import SMOTENC

features_rs, target_rs = SMOTENC(categorical_features = [0,3,5,6],sampling_strategy='auto',n_jobs=-1).fit_sample(features, target)
print(pd.Series(target_rs).value_counts()) #Preview synthetic sample class distribution
```

```python
data_rs = pd.DataFrame(features_rs)
```

```python
data_rs.head()
```

```python
data_rs.info()
```

```python
data_rs[20] = target_rs
```

```python
data_rs.head()
```

```python
data_rs = data_rs.drop([21], axis=1)
```
